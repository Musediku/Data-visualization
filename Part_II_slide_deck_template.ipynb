4. Start Station ID

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('Bike_df.csv')
```

## Does the trip duration taken determinded by if a user is a subscriber or a customer?

> The data shows that  the trip duration of the trip affected by  if a user is a subscriber or just a customer. Customers ( more than 1200 seconds) have two times the trip duration lenght of subscribers (600 seconds) This means Customer users spend more time on the ride than Subriber type of users.

```python
# Average trip duration mean plot
user_type_duration = df[['user_type','duration_sec']]
sb.barplot(x='user_type', y='duration_sec', data=user_type_duration)
plt.title("duration by User Types")
plt.ylabel("Trip Duration (sec)")
plt.show();
```

##  How is the distance covered affected by the duration???

> The visualization shows the distribution of rides based on their duration in secs and distace in kms.

> To better understand distribution we removed the outliers that had a distance of more than 20 kms.

> The visualization shows most of the rides have short duration (under 30000 sec) and short length (under 10 kms). Which give a clear indication of that usage pattern of bike riders. This means  users use it for shorter rides within city limits.



```python

# Lets drop the outliers and create a plot 
clean_distance = df[df['distance'] < 20 ]
plt.figure(figsize=[11, 6])
plt.scatter(data = clean_distance, x = 'distance', y = 'duration_sec')
plt.title('Duration and Distance', fontsize=16)
plt.xlabel('Distance in kms',fontsize=12)
plt.ylabel('Duration in seconds',fontsize=12);
```

## Trips per Gender

> Most of the trips were made by clients who identify as a male, then females, and only negligible numbers of the trips were of other users.



```python
plt.figure(figsize=(6, 6))
order_plot = df.member_gender.value_counts().index

# plotting
fig = sb.countplot(data=df, x='member_gender', order=order_plot)
fig.set_title('Trips per Gender')
fig.set_xlabel('Genders')
fig.set_ylabel('Numbers of trips')

plt.show()
```
